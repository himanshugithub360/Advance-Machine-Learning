### Handling the missing values in different features

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv("E:/Python and ML/Dataset/covid_toy.csv")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df.drop(columns = ["has_covid"]), df["has_covid"], test_size=0.2, random_state=42)
```

```python
X_train
```

```python
# adding simple imputer for numerical columns
si = SimpleImputer()
X_train_fever = si.fit_transform(X_train[["fever"]])

# also on the test data
X_test_fever = si.transform(X_test[["fever"]])
```

```python
X_train_fever.shape
```

```python
# Ordinal Encoding -> cough
oe = OrdinalEncoder(categories=[["Mild", "Strong"]])
X_train_cough = oe.fit_transform(X_train[["cough"]])

# also on the test data
X_test_cough = oe.fit_transform(X_test[["cough"]])

```

```python
X_train_cough.shape
```

```python
X_test_cough.shape
```

```python
# OneHotEncoding -> gender, city
ohe = OneHotEncoder(drop = 'first',sparse_output=False)
X_train_gender_city = ohe.fit_transform(X_train[["gender","city"]])

# also on the test data
X_test_gender_city = ohe.fit_transform(X_test[["gender","city"]])
```

```python
X_test_gender_city
```

```python
X_test_gender_city.shape
```

```python
# Extracting Age
X_train_age = X_train.drop(columns = ["fever","cough","gender","city"]).values

#also on the test data
X_test_age = X_test.drop(columns = ["fever","cough","gender","city"]).values
```

```python
X_train_age.shape
```

```python
X_train_transformed = np.concatenate((X_train_age, X_train_fever,  X_train_gender_city, X_train_cough), axis = 1)

#also on the test data

X_test_transformed = np.concatenate((X_test_age,X_test_fever,X_test_gender_city,X_test_cough),axis=1)
```

## Now with columnTransformer

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers =[
    ("tnf1", SimpleImputer(),["fever"]),
    ("tnf2", OrdinalEncoder(categories=[["Mild", "Strong"]]),["cough"]),
    ("tnf3", OneHotEncoder(sparse_output =False,drop = 'first'),["gender","city"])
],remainder = 'passthrough') 
```

```python
df = transformer.fit_transform(X_train) 
```
