### Working with Dates

```python
# converting to datetime datatype

date["date"] = pd.to_datetime(date["date"])
```

```python
date.info()
```

### 1. Extract year

```python
date["date_year"] = date["date"].dt.year
date.sample(5)
```

### 2. Extract Month

```python
date["date_month_no"] = date["date"].dt.month
date.head()
```

```python
date["date_month_name"] = date["date"].dt.month_name()
date.head()
```

```python
date.head()
```

### 3. Extract Days

```python
date["date_day"] = date["date"].dt.day
date.head()
```

```python
# day of the week
date["date_dow"] = date["date"].dt.dayofweek
date.head()
```

```python
# day of week - name

date["date_dow_name"] = date["date"].dt.day_name()
date.head()
```

```python
date.drop(columns =["product_id","city_id","orders"], inplace = True)
```

```python
date.head()
```

```python
# is weekend?
date["date_is_weekend"] = np.where(date["date_dow_name"].isin(["Saturday","Sunday"]),1,0)
date.head()
```

### 4. Extract week of the year

```python
date["date_week"] = date["date"].dt.weekday
date.head()
```

### 5.Extract Quarter

```python
date["quarter"] = date["date"].dt.quarter
date.head()
```

### 6. Extract Semester

```python
date["semester"] = np.where(date["quarter"].isin([1,2]),1,2)
date.head()
```

### 7. Extract Time elapsed between dates

```python
import datetime
```

```python
today = datetime.datetime.today()
today
```

```python
today - date["date"]
```

```python
(today - date["date"]).dt.days
```

```python
# Months passed

np.round((today - date["date"])/np.timedelta64(1,'m'),0)
```

```python
time.info()
```

```python
# converting to datetime datatype
time["date"] = pd.to_datetime(time["date"])
```

```python
time.info()
```

```python
time['hour'] = time['date'].dt.hour
time['min'] = time['date'].dt.minute
time['sec'] = time['date'].dt.second

time.head()
```

### Extract the time part

```python
time["time"] = time["date"].dt.time
time.head()
```

### Time difference

```python
today - time["date"]
```

```python
# in seconds

(today - time["date"])/np.timedelta64(1,'s')
```

```python
# in minutes

(today - time["date"])/np.timedelta64(1,'m')
```

```python
# in hours

(today - time['date'])/np.timedelta64(1,'h')
```
