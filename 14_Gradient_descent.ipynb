## Gradient Descent

```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,Y = make_regression(n_samples=4, n_features= 1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y)
```

```python
# lets apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,Y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,Y)
plt.plot(X, reg.predict(X), color = "red")
```

```python
# Lets apply Gradient Descent assuming slope is constant m = 78.35
# and let's assumne that the starting value fir intercept b = 0
y_pred = ((78.35 * X) + 0).reshape(4)
y_pred
```

```python
plt.scatter(X,Y)
plt.plot(X, reg.predict(X), color = "red", label = "OLS")
plt.plot(X, y_pred, color = "green", label = "b = 0")
plt.legend()
```

```python
m = 78.35
b = 100

loss_slope = -2 * np.sum(Y - m * X.ravel() - b)
loss_slope
```

```python
# lets take learning rate = 0.1

lr = 0.1
step_size = loss_slope*lr
step_size
```

```python
# calculating the new intercept

b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration -2
loss_slope = -2 * np.sum(Y - m * X.ravel() - b)
loss_slope 
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)
```

```python
plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color="#1300a6",label='b = {}'.format(b))
plt.plot(X,y_pred1,color="#01FF1F",label='b = 40.92' )
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 3
loss_slope = -2 * np.sum(Y - m*X.ravel() - b)
loss_slope
     
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
     
```

```python
y_pred3 = ((78.35 * X) + b).reshape(4)
```

```python
plt.figure(figsize=(15,15))
plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred3,color="#abd539",label='b = {}'.format(b))
plt.plot(X,y_pred2,color="#0B3DF1",label='b = 29.75')
plt.plot(X,y_pred1,color="#00F94F",label='b = 40')
plt.plot(X,y_pred,color="#E410D2",label='b = 0')
plt.legend()
plt.show()
```

```python
b = -100
m = 78.35
lr = 0.01

epochs = 100

for i in range(epochs):
  loss_slope = -2 * np.sum(Y - m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m * X + b

  plt.plot(X,y_pred)

plt.scatter(X,Y)
```
